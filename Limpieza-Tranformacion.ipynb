##### Transformacion de datos.

```python
df_sin_nulos['salary']
```

###### Esta funcion toma los valores de la columna salary y los convierte en valores numericos. El resultado se almacena en una nueva columna 'salary_numeric'

```python
df_sin_nulos['salary_numeric'] = df_sin_nulos['salary'].replace('[^\d.]', '', regex=True).str.extract('(\d+)').astype(float)
df_sin_nulos['salary_numeric']
```

```python
df_sin_nulos['salary_numeric'].describe()
```

```python
#Crea boxplot con la columna de salarios.
plt.boxplot(df_sin_nulos['salary_numeric'])

#Etiquetas y titulo.
plt.xlabel('Salarios')
plt.ylabel('Valores USD')
plt.title('Boxplot')

#Muestra el grafico
plt.show()
```

```python
#Crear histograma
plt.hist(df_sin_nulos['salary_numeric'], bins=30, color='yellow', alpha=0.7)

#Etiquetas y titulo
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma con valores de salarios')

#Mostrar histograma
plt.show()

```

```python
#Filtro aquellos salarios mayores o iguales a 20000 ya que el archivo csv contiene
#salarios mensuales y anules.
df_sin_nulos = df_sin_nulos[df_sin_nulos['salary_numeric']>= 20000]
```

```python
#Crea boxplot
plt.boxplot(df_sin_nulos['salary_numeric'])

#Etiquetas y titulo
plt.xlabel('Salarios')
plt.ylabel('Valores en USD')
plt.title('Boxplot')

#Muestra grafico
plt.show()
```

```python
#Crear histograma 
plt.hist(df_sin_nulos['salary_numeric'], bins=30, color='pink', alpha=0.7)

#Etiquetas y titulo
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma con valores de salarios mayor a 20000')

#Mostrar histograma
plt.show()
```

#### Metodo de reemplazo de valores nulos con la media de la columna

```python
#Rellena valores nulos con un valor especifico (el promedio)
df_con_relleno = df_sin_nulos.fillna(df_sin_nulos['salary_numeric'].mean())
```

```python
#Elimina la columna salary.
df_con_relleno.drop('salary', axis=1, inplace=True)
df_con_relleno
```

#### Agrupar y agregar datos

```python
#En un nuevo DF agreupar por titulo y calcular el promedio y la mediana del 
#salario por cada titulo en el archivo csv.
df_grouped = df_con_relleno.groupby('title')['salary_numeric'].agg(['mean', 'median', 'count']).reset_index()
```

```python
#Renombro columnas
df_grouped.columns = ['title', 'salary_mean', 'salary_median', 'job_count']

df_grouped
```

```python
#Redondear valores.
df_grouped['salary_mean'] = df_grouped['salary_mean'].round(0).astype(int)
df_grouped['salary_median'] - df_grouped['salary_median'].round(0).astype(int)

df_grouped
```

#### Exportar CSV.

```python
ruta_archivo_csv = "export-linkedin-jobs-usa"
df_grouped.to_csv(ruta_archivo_csv)
```
